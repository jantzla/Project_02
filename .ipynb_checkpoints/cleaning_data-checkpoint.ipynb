# Mental Data

```python
df_mental = pd.read_csv(r"C:\Users\ljant\Desktop\Ironhack\Projects\Project_2\mental_python.csv")
df_mental
# 43.928 x 11
```

```python
# Dropping duplicates if present
df_mental.drop_duplicates()
```

```python
# Combinations of 7s and 9s are used when data is not applicable or when the patient refused to answer
df_mental.replace({7: np.nan, 77: np.nan, 9: np.nan, 99: np.nan}, inplace=True)
```

```python
# Assessing NaNs
sum_na = df_mental.isna().sum()
print (sum_na)
```

```python
# Adding column with the sum of the PHQ-Score
df_mental['total_score'] = df_mental[['q1', 'q2', 'q3', 'q4', 'q5', 'q6', 'q7', 'q8', 'q9']].apply(sum, axis=1)
```

```python
bins = [-1, 4, 9, 14, 19, 27]
labels = ['minimal', 'mild', 'moderate', 'moderately_severe', 'severe']
df_mental['depression_level'] = pd.cut(df_mental['total_score'], bins=bins, labels=labels)
df_mental
```

# Joining the tables

```python
df_total = pd.merge(df_mental, df_demo, how = 'left', on = 'patient_id')
df_total = pd.merge(df_total, df_diet, how = 'left', on = 'patient_id')
df_total
# 43.928 x 51
```

# Final filtering before further analysis

```python
# Age between 18-150 (rule by questionaire)
df_final = df_total[(df_total['age'] >= 18) & (df_total['age'] <= 150)]

# Only using data with complete mental questionaire
df_final = df_total.dropna(subset=['q1','q2','q3','q4','q5','q6','q7','q8','q9'])

df_final
# 39.467 x 51
```
