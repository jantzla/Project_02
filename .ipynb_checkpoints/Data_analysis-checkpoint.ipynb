# Dealing with outliers

```python
df_model.describe()
```

```python
def handling_outliers(data, column, std_devs=3):
    mean = data[column].mean()
    std_dev = data[column].std()

    lower_limit = mean - (std_devs * std_dev)
    upper_limit = mean + (std_devs * std_dev)

    data[column] = np.where(data[column] < lower_limit, lower_limit, data[column])
    data[column] = np.where(data[column] > upper_limit, upper_limit, data[column])
    
    return data

columns = ['protein', 'fiber', 'sattuarated_fatty_acids',
       'monounsaturated_fatty_acids', 'polyunsaturated_fatty_acids',
       'cholesterol', 'carotene', 'vitamin_a', 'vitamin_b1', 'vitamin_b2',
       'vitamin_b6', 'vitamin_b12', 'vitamin_c', 'vitamin_e', 'folate',
       'calcium', 'phosporus', 'magnesium', 'iron', 'zinc', 'copper', 'sodium',
       'potassium', 'selenium', 'caffeine', 'alcohol', 'water', 'total_sugars',
       'vitamin_d', 'vitamin_k', 'choline']

df_adj = handling_outliers(df_model, columns)
df_adj
```

# Statistical hypothesis testing
"Do participants classified as 'depressed' have significant differences in their nutrion intake profile compared to participants  that are classified as 'non_depressed'?"

```python
# 1. Bins for depressed/not_depressed
bins = [-1, 9, 27]
labels = ['not_depressed', 'depressed']
df_adj['depression_level'] = pd.cut(df_adj['total_score'], bins=bins, labels=labels)
df_adj
```

```python
# 2. Statistical tests
# Split the dataset
depressed = df_adj[df_adj['depression_level'] == 'depressed']
not_depressed = df_adj[df_adj['depression_level'] == 'not_depressed']

variables = ['age', 'gender','protein', 'fiber', 'sattuarated_fatty_acids', 'monounsaturated_fatty_acids', 'polyunsaturated_fatty_acids', 
             'cholesterol', 'carotene', 'vitamin_a', 'vitamin_b1', 'vitamin_b2', 'vitamin_b6', 'vitamin_b12', 
             'vitamin_c', 'vitamin_e', 'folate', 'calcium', 'phosporus', 'magnesium', 'iron', 'zinc', 
             'copper', 'sodium', 'potassium', 'selenium', 'caffeine', 'alcohol', 'water', 'total_sugars', 
             'vitamin_d', 'vitamin_k', 'choline']

# Perform Mann-Whitney U tests [no (!) normal destribution of the data]
results = {}
for variable in variables:
    u_stat, p_val = stats.mannwhitneyu(not_depressed[variable], depressed[variable], alternative='two-sided')

    results[variable] = {'u_stat': u_stat, 'p_val': p_val}

df_results = pd.DataFrame(results).T

alpha = 0.05

df_results['significant'] = df_results['p_val'] < alpha

df_results
```
